```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv('./data/gro_train.csv').iloc[:, 1:]
valid = pd.read_csv('./data/gro_test.csv').iloc[:, 1:]
```

```python
valid
```

```python
train
```

```python
valid
```

```python
time_steps = 100
ts_train_len = len(train)
ts_valid_len = len(valid)
for_periods = 14
```

```python
X_train = []
y_train = []
for i in range(time_steps, ts_train_len-for_periods):
    X_train.append(train.iloc[i-time_steps:i])
    y_train.append(train.iloc[i:i + for_periods, 0])
X_train, y_train = np.array(X_train), np.array(y_train)
```

```python
X_valid = []
y_valid = []
for i in range(time_steps, ts_valid_len-for_periods):
    X_valid.append(valid.iloc[i-time_steps:i])
    y_valid.append(valid.iloc[i:i + for_periods, 0])
X_valid, y_valid = np.array(X_valid), np.array(y_valid)
```

```python
X_train.shape, y_train.shape
```

```python
X_valid.shape, y_valid.shape
```

```python
from tensorflow import keras
cell = 100
model = keras.models.Sequential([
        keras.layers.LSTM((cell), input_shape=[None, 3], return_sequences=True),
        keras.layers.SimpleRNN((20)),
        keras.layers.Dense(for_periods)
])
```

```python
model.compile(loss=keras.losses.MeanSquaredError(), optimizer=keras.optimizers.Adam())
```

```python
history = model.fit(X_train, y_train, epochs=100, validation_data=(X_valid, y_valid))
```

```python
history = model.fit(X_train, y_train, initial_epoch=120, epochs=140, validation_data=(X_valid, y_valid))
```

```python
import matplotlib.pyplot as plt
y_vloss = history.history['val_loss']
y_loss = history.history['loss']

x_len = np.arange(len(y_loss))
plt.plot(x_len, y_vloss, marker='.', c='red', label="Validation-set Loss")
plt.plot(x_len, y_loss, marker='.', c='blue', label="Train-set Loss")
plt.title(f'cell : {cell}, time_steps : {time_steps}, periods : {for_periods}')
plt.legend(loc='upper right')
plt.grid()
plt.xlabel('epoch')
plt.ylabel('loss')
plt.show()
```

### 한개 예측

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (10, 4))
plt.plot(valid.iloc[time_steps:-for_periods, 0].values)
plt.plot(model.predict(X_valid)[:])

plt.legend(['True', 'Predict'])
plt.show()
```

### 여러개 예측

```python
i = 2
plt.figure(figsize= (15, 4))
plt.plot(np.arange(time_steps + for_periods), valid.iloc[i:time_steps + for_periods+i, 0].values)
plt.plot(np.arange(time_steps, time_steps + for_periods), valid.iloc[i+time_steps:time_steps + for_periods+i, 0].values)
plt.plot(np.arange(time_steps, time_steps + for_periods) , model.predict(np.expand_dims(X_valid[i], axis = 0)).squeeze())
plt.vlines(time_steps, 0, 1, color = 'gray', linestyle = '--')
plt.vlines(time_steps + 7, 0, 1, color = 'red', linestyle = '--')
plt.ylim(0,valid.iloc[:time_steps + for_periods, 0].values.max() + 0.2)
plt.title(f'cell : {cell}, time_steps : {time_steps}, periods : {for_periods}')
plt.grid()
plt.legend(['period', 'True', 'Predict'])
plt.show()
```
