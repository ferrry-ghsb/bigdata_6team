```python
import pandas as pd
import numpy as np
%load_ext autoreload
%autoreload 2
```

```python
train = pd.read_csv('./data/bea_train.csv').iloc[:, 1:]
valid = pd.read_csv('./data/bea_test.csv').iloc[:, 1:]
```

```python
valid
```

```python
train
```

```python
train.iloc[1:1+1, :].values
```

```python
valid
```

```python
time_steps = 100
ts_train_len = len(train)
ts_valid_len = len(valid)
output = 2
for_periods = 14
```

```python
X_train = []
y_train = []
for i in range(time_steps, ts_train_len-1):
    X_train.append(train.iloc[i-time_steps:i])
    y_train.append(train.iloc[i:i+1, [0,2]].values.squeeze().tolist())
X_train, y_train = np.array(X_train), np.array(y_train)
```

```python
X_valid = []
y_valid = []
for i in range(time_steps, ts_valid_len-1):
    X_valid.append(valid.iloc[i-time_steps:i])
    y_valid.append(valid.iloc[i:i+1, [0,2]].values.squeeze().tolist())
X_valid, y_valid = np.array(X_valid), np.array(y_valid)
```

```python
X_train.shape, y_train.shape
```

```python
X_valid.shape, y_valid.shape
```

```python
X_valid[0].shape
```

```python
from tensorflow import keras

model = keras.models.Sequential([
        keras.layers.LSTM(100, input_shape=[None, X_train.shape[2]]),
        
        keras.layers.Dense(output)
])
```

```python
optimizer=keras.optimizers.Adam()
```

```python
optimizer.learning_rate = 0.01
```

```python
optimizer.learning_rate
```

```python
model.compile(loss=keras.losses.MeanSquaredError(), optimizer=keras.optimizers.Adam())
```

```python
history = model.fit(X_train, y_train, epochs=100, validation_data=(X_valid, y_valid))
```

```python
keras.optimizers.Adam().learning_rate
```

```python
import matplotlib.pyplot as plt
y_vloss = history.history['val_loss']
y_loss = history.history['loss']

x_len = np.arange(len(y_loss))
plt.plot(x_len, y_vloss, marker='.', c='red', label="Validation-set Loss")
plt.plot(x_len, y_loss, marker='.', c='blue', label="Train-set Loss")
plt.title(f'time_steps : {time_steps}, periods : {for_periods}')
plt.legend(loc='upper right')
plt.grid()
plt.xlabel('epoch')
plt.ylabel('loss')
plt.show()
```

```python
def period_prediction(model, X_valid, on_promotion, for_periods):
    predict = []
    X_pred = X_valid.copy()
    for i in range(for_periods):
        pred = model.predict(np.expand_dims(X_pred, axis = 0))
        
        predict += pred.tolist()
        X_pred = np.append(X_pred[1:], np.expand_dims(np.insert(pred, 1, on_promotion[i]), axis = 0), axis = 0)
    
    return np.array(predict)
```

```python
i = 0
for_periods = 14
promotion = valid.iloc[i+time_steps:time_steps + for_periods+i, 1].values
predict = period_prediction(model, X_valid[i], promotion, for_periods)

plt.figure(figsize= (15, 4))

# period
plt.plot(np.arange(time_steps + for_periods), valid.iloc[i:time_steps + for_periods+i, 0].values)

# True
plt.plot(np.arange(time_steps, time_steps + for_periods), valid.iloc[i+time_steps:time_steps + for_periods+i, 0].values)

# predict
plt.plot(np.arange(time_steps, time_steps + for_periods) , predict[:, 0])
plt.plot(np.arange(time_steps, time_steps + for_periods) , promotion, alpha = 0.4)

plt.vlines(time_steps, 0, 1, color = 'gray', linestyle = '--')
plt.vlines(time_steps + 7, 0, 1, color = 'red', linestyle = '--')
plt.ylim(valid.iloc[:time_steps + for_periods, 0].values.min() - 0.1,valid.iloc[:time_steps + for_periods, 0].values.max() + 0.2)
plt.title(f', time_steps : {time_steps}, periods : {for_periods}')
plt.grid()
plt.legend(['period', 'True', 'Predict', 'Promotion'])
plt.show()
```
