# 현재는 안 쓰는 원핫인코딩

```python
#원핫인코딩을 위한 함수 정의
def dummy_data(data, columns):
    for column in columns:
        data = pd.concat([data, pd.get_dummies(data[column], prefix = column)], axis=1)
        data = data.drop(column, axis=1)
    return data
```

```python
trans_cols = ["store_nbr", "family", "city", "state", "type_x", "cluster", "type_y", "locale", "locale_name", "description",	"transferred"]
```

```python
train_trans = dummy_data(train_cp, trans_cols)
```

```python
#원핫인코딩 잘 진행됐는지 확인
train_trans
```
