```python
hol = hol[41:328]
```

```python
#train과 store을 store nber 기준으로 합병. 
#합병된 df이름은 train_cp
train_cp = train.merge(stores, how = 'left', left_on='store_nbr', right_on='store_nbr')
```

```python
train_cp
```

```python
#train+store+oil
train_cp = train_cp.merge(oil, how = 'left', left_on='date', right_on='date')
```

```python
train_cp
```

```python
#train+store+oil+holiday
train_cp = train_cp.merge(hol, how = 'left', left_on='date', right_on='date')
```

```python
train_cp
```

```python
train_cp.dtypes
```

```python
#object는 one hot encoding으로 바꾸는 게 좋을 듯
#dcoilwtico 는 NULL 값 채워야 함.
train_cp.isnull().sum()
```

```python
#dcoilwtico의 null값을 뒷값으로 대체(첫행부터 값이 존재하지 않아서 뒷값으로 대체함)
train_cp['dcoilwtico'].fillna(method='bfill', inplace=True)
```
